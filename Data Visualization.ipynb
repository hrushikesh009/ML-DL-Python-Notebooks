Definig axes,ticks, grids

```
x = range(0,10)
y = [5,6,3,4,5,6,2,8,9,10]

fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])

ax.grid()
ax.plot(x,y)
```

setting Limits and ticks on x,y Axis

```
fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])

ax.set_xlim([1,9])
ax.set_ylim([0,5])

ax.set_xticks([0,3,7,5])#to be show along x-axis
ax.set_yticks([2,8,4,1])#to be show along x-axis
ax.grid()

ax.plot(x,y)

```
