```
import turicreate as tc
```

```
sales = tc.SFrame('/content/drive/My Drive/home_data.sframe')
sales
```

```
training_data,test_data = sales.random_split(.8,seed=0)
```

```
input_feature = training_data['sqft_living']
output = training_data['price']
```

```
input_feature_bedroom = training_data['bedrooms']
```

```
def simple_linear_regression(input_feature, output):
    # compute the number of data points 
    N = len(training_data)
    print(N)
    # compute the sum of input_feature and output
    X = input_feature.sum()
    Y = output.sum()
    print(X)
    print(Y)
    # compute the product of the output and the input_feature and its sum
    product = input_feature*output
    Z = product.sum()
    print(Z)
    
    # compute the squared value of the input_feature and its sum
    product_1 = input_feature*input_feature
    W = product_1.sum()
    print(W)

    
    # use the formula for the slope
    numerator = (Z) - (1/N) * (X*Y)
    denominator = (W) - (1/N) * (X*X)
    slope = numerator/denominator

    
    # use the formula for the intercept
    intercept = (1/N)*Y - slope * (1/N)*X 
    return (intercept, slope)
```

```
simple_linear_regression(input_feature,output)
```

```
simple_linear_regression(input_feature_bedroom,output)
```

```
def get_regression_predictions(input_feature, intercept, slope):
    # calculate the predicted values:
    
    predicted_values = tc.SArray(intercept + slope*input_feature)
    
    return predicted_values
```

```
def get_residual_sum_of_squares(input_feature, output, intercept,slope):
   predicted_values = tc.SArray(intercept + slope*input_feature)
   value = output - predicted_values
   values = value*value
   RSS = values.sum()
   return RSS

    
```
