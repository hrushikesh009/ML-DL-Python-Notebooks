```
import turicreate as tc
```

```
image_train = tc.SFrame('/content/drive/My Drive/image_train_data/')
image_test = tc.SFrame('/content/drive/My Drive/image_test_data/')
```

```
dog_image_train = image_train[image_train['label'] == 'dog']
dog_image_test =  image_test[image_test['label'] == 'dog']
```

```
cat_image_train = image_train[image_train['label'] == 'cat']
cat_image_test =  image_test[image_test['label'] == 'cat']
```

```
automobile_image_train = image_train[image_train['label'] == 'automobile']
automobile_image_test =  image_test[image_test['label'] == 'automobile']
```

```
bird_image_train = image_train[image_train['label'] == 'bird']
bird_image_test =  image_test[image_test['label'] == 'bird']
```

```
dog_model = tc.nearest_neighbors.create(dog_image_train,features=['deep_features'],label ='id')
```

```
cat_model = tc.nearest_neighbors.create(cat_image_train,features=['deep_features'],label ='id')
```

```
bird_model = tc.nearest_neighbors.create(bird_image_train,features=['deep_features'],label ='id')
```

```
automobile_model = tc.nearest_neighbors.create(automobile_image_train,features=['deep_features'],label ='id')
```

```
dog_cat_neighbors = cat_model.query(dog_image_test, k=1)
```

```
dog_bird_neighbors = bird_model.query(dog_image_test, k=1)
```

```
dog_dog_neighbors = dog_model.query(dog_image_test, k=1)
```

```
dog_automobile_neighbors = automobile_model.query(dog_image_test, k=1)
```

```
dog_distances = tc.SFrame({'dog-automobile': dog_automobile_neighbors['distance'],'dog-bird': dog_bird_neighbors['distance'],'dog-cat': dog_cat_neighbors['distance'],'dog-dog': dog_dog_neighbors['distance']})
```

```
dog_distances
```

```
i = 0
count = 0
for i in range(1000):
  if dog_distances['dog-dog'][i] < dog_distances['dog-automobile'][i]:
    if dog_distances['dog-dog'][i] < dog_distances['dog-bird'][i]:
      if dog_distances['dog-dog'][i] < dog_distances['dog-cat'][i]:
        count = count + 1
        
print(count)
    
      
```
