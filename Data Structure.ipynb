# Stack 

```
class Stack():
    def __init__(self):
        self.items=[]
    def push(self ,item):
        self.items.append(item)
    def pop(self):
        return self.items.pop()
    def flush():
        self.items = []
    def is_empty(self):
        return self.items == []
    def peek(self):
        if not self.is_empty():
            return self.items[-1]
    def get_stack(self):
        return self.items
    
    def __len__(self):
        return self.size()

    def size(self):
        return len(self.items)

def is_match(p1,p2):
    if p1 == "(" and p2 == ")" :
        return True
    elif p1 == "{" and p2 == "}":
        return True
    elif p1 == "[" and p2 == "]":
        return True
    else:
        return False
def is_paren_balanced(paren_string):
    index = 0
    s = Stack()
    is_balanced=True
    
    while index < len(paren_string) and is_balanced:
        paren = paren_string[index]
        if paren in "([{":
            s.push(paren)
        else:
            if s.is_empty():
                is_balanced = False
            else:
                top = s.pop()
                if not is_match(top, paren):
                    is_balanced = False
        index += 1
    if s.is_empty() and is_balanced:
        return True
    else:
        return False 
    
print(is_paren_balanced('({[{}][]})')) 

```

# Integer to Binary Conversion

```
def int_to_bin(num):
    s1 = Stack()
    
    while num > 0:
        rem = num % 2
        s1.push(rem)
        num = num // 2
    
    bin_num = ""
    while not s1.is_empty():
        bin_num += str(s1.pop())
        
    return bin_num

print(int_to_bin(20))  
```

```
data = [2,4,5,6,7,8,9,10,23,45,66,88,90]
target = 8
def Binary_search(data,target):
  low = 0
  high = len(data) - 1

  while(low <= high):
    mid = (low + high) // 2
    if target == data[mid]:
      return True
    elif target < data[mid]:
      high = mid - 1
    else :
      low = mid + 1
  return False

Binary_search(data,target)
```

# Binary Search Recursively

```
def Binary_search_recursive(data,target,low,high):
  if low > high:
    return False
  else:
    mid = (low + high) // 2
    if target == data[mid]:
      return True
    elif target < data[mid]:
      return Binary_search_recursive(data, target, low,mid - 1)
    else :
      return Binary_search_recursive(data,target, mid + 1 ,high)
  

Binary_search_recursive(data,target,0,len(data)-1)
```

#Interpolation Search


```
def interpolation(a,n,x):
  l = 0
  h = n-1
  while l<=h and x >= a[l] and x <= a[h]:
    if l == h:
      if a[l] ==x:
        return l
      return -1
    pos  = l + int(((float(h - l) / 
            ( a[h] - a[l])) * ( x - a[l])))
    
    if a[pos] == x:
      return pos
    if a[pos] < x:
      l = pos + 1
    else:
      h = pos-1
  return -1
data = [2,4,5,6,7,8,9,10,23,45,66,88,90]
target = 8
interpolation(data,len(data),target)
```

#Insertion Sort

```
#insertion sort
def insertion_sort(value):
  n = len(value)
  for i in range(1,n):
    key = value[i]

    j = i-1 
    while j >= 0 and key < value[j]:
      value[j+1] = value[j]
      j -= 1
    value[j+1] = key
  return value
arr = [100,45,7,89,23,56,89,112,32,56,8,0]
print(arr)
print(insertion_sort(arr))
    
```

#Shell Sort

```
def shellSort(arr):
    n = len(arr) 
    gap = n//2
    while gap > 0:
      for i in range(gap,n): 
            temp = arr[i] 
            j = i 
            while  j >= gap and arr[j-gap] >temp: 
                arr[j] = arr[j-gap] 
                j -= gap
            arr[j] = temp
      gap //= 2
arr = [ 10,7,8,9,1,5] 
shellSort(arr) 
print(arr) 

```

```
def partition(arr,low,high):
  i = low-1
  p = arr[high]

  for j in range(low,high):
    if arr[j] < p:
      i += 1
      arr[i],arr[j] = arr[j],arr[i]
    
  
  arr[i+1],arr[high] = arr[high],arr[i+1]
  return ( i+1 )

def quickSort(arr,low,high): 
    if low < high: 
  
         
        pi = partition(arr,low,high) 
  
        
        quickSort(arr, low, pi-1) 
        quickSort(arr, pi+1, high) 

arr = [10, 7, 8, 9, 1, 5] 
n = len(arr) 
quickSort(arr,0,n-1) 
print(arr)
```

#heap Sort

```
def heapify(arr, n, i):
   
    largest = i 
    l = 2 * i + 1
    r = 2 * i + 2
   
  
    if l < n and arr[i] < arr[l]: 
        largest = l 
  
    if r < n and arr[largest] < arr[r]: 
        largest = r 
  
     
    if largest != i: 
        arr[i],arr[largest] = arr[largest],arr[i] 
  
       
        heapify(arr, n, largest) 
  

def heapSort(arr): 
    n = len(arr) 
  
     
    for i in range(n//2 - 1, -1, -1): 
        heapify(arr, n, i) 
  
    for i in range(n-1, 0, -1): 
        arr[i], arr[0] = arr[0], arr[i] 
        heapify(arr, i, 0) 
  

arr = [ 12, 11, 13, 5, 6, 7] 
heapSort(arr) 
n = len(arr) 
print(arr)
```

#BiggerIsGreater on Hackerrank

```
def biggerIsGreater(arr):
    value = arr
    if len(arr) == 1:
        return "no answer" 
    if len(arr) == 2:
        if arr[0] != arr[1]:
            arr = list(arr)
            arr.reverse()
            if ''.join(arr) > value:
                return "".join(arr)
            else:
                return "no answer"
        else:
            return "no answer"
    else:
        arr = list(arr)
        largestx = -1
        largesty = -1
        for i in range(len(arr)-1):
            if arr[i] < arr[i+1]:
                largestx = i
       
        for j in range(len(arr)):
            if arr[largestx] < arr[j]:
                largesty = j
        
            
        arr[largestx],arr[largesty] = arr[largesty],arr[largestx]
        
        endarr = arr[largestx+1:]
        if "".join(arr[:largestx+1]) + "".join(reversed(endarr)) > value:
            return "".join(arr[:largestx+1]) + "".join(reversed(endarr))
        else:
            return "no answer"
        
value = []   
for _ in range(int(input())):
    value.append(input())
for x in value:
    print(biggerIsGreater(x))
        
```
