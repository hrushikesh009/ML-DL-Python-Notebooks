```python
fashion_data = tc.SFrame.read_json('/content/drive/My Drive/meta_AMAZON_FASHION.json',orient='lines')
```

```python
fashion_data.remove_columns(['tech1','also_buy','also_view','tech2','feature','category','similar_item','details','main_cat','date','fit'],inplace = True)
fashion_data.rename({'asin':'ProdID'},inplace=True)
```

```python
fashion_data = fashion_data[fashion_data['price'] != '']
fashion_data = fashion_data[fashion_data['description'] != []]
fashion_data = fashion_data[fashion_data['image'] != []]
```

```python
fashion_data
```

```python
fashion_data['rating'] = 0
fashion_data['Sub_Category'] = '[]'
fashion_data['Rank'] = '[]'
```

```python
def change_description(ProdID):
  return fashion_dataset[fashion_dataset['ProdID'] == ProdID]['description'][0]
def change_Rank(ProdID):
  return fashion_dataset[fashion_dataset['ProdID'] == ProdID]['Rank'][0]
def change_rank(ProdID):
  return fashion_dataset[fashion_dataset['ProdID'] == ProdID]['rank'][0]
def change_Sub_Category(ProdID):
  return fashion_dataset[fashion_dataset['ProdID'] == ProdID]['Sub_Category'][0]
def change_price(ProdID):
  return fashion_dataset[fashion_dataset['ProdID'] == ProdID]['price'][0]
def change_brand(ProdID):
  return fashion_dataset[fashion_dataset['ProdID'] == ProdID]['brand'][0]
def change_title(ProdID):
  return fashion_dataset[fashion_dataset['ProdID'] == ProdID]['title'][0]
def change_rating(ProdID):
  return fashion_dataset[fashion_dataset['ProdID'] == ProdID]['rating'][0]

```

```python
fashion_data['description'] = fashion_data['ProdID'].apply(change_description)
fashion_data['Rank'] = fashion_data['ProdID'].apply(change_Rank)
fashion_data['rank'] = fashion_data['ProdID'].apply(change_rank)
fashion_data['Sub_Category'] = fashion_data['ProdID'].apply(change_Sub_Category)
fashion_data['price'] = fashion_data['ProdID'].apply(change_price)
fashion_data['brand'] = fashion_data['ProdID'].apply(change_brand)
fashion_data['title'] = fashion_data['ProdID'].apply(change_title)
fashion_data['rating'] = fashion_data['ProdID'].apply(change_rating)
```

```python
fashion_data = tc.SFrame.read_csv('/content/drive/My Drive/product_dataset.csv')
```

```python
value = fashion_data['ProdID'].unique()
```

```python
def change_prodid(ProdID):
  if ProdID in list(value):
    return ProdID
  return None
```

```python
fashion['ProdID'] = fashion['ProdID'].apply(change_prodid)
fashion = fashion[fashion['ProdID'] != None]
```

#Recommender Model Training

```python
train,test = tc.recommender.util.random_split_by_user(fashion,user_id='UserID',item_id='ProdID',random_seed= 8)
```

```python
Popularity_model = tc.recommender.popularity_recommender.create(train,user_id ='UserID',item_id = 'ProdID',target = 'Rating')
```

```python
Popularity_model.evaluate(test)
```

```python
Popular_Products = list(Popularity_model.recommend([train["UserID"][0]])['ProdID'])
```

```python
Popular_Products
```

```python
factorization_recommender_model_adagrad = tc.recommender.factorization_recommender.create(train,user_id = 'UserID',item_id = 'ProdID',target= 'Rating',max_iterations=200,
                                                          regularization=1e-05,solver = 'adagrad',linear_regularization = 1e-05)
```

```python
factorization_recommender_model_adagrad.evaluate(test)
```

```python
factorization_recommender_model_adagrad.recommend([test['UserID'][100]])
```
