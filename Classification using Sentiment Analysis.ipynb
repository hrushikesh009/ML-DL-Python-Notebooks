```
selected_words = ['awesome', 'great', 'fantastic', 'amazing', 'love', 'horrible', 'bad', 'terrible', 'awful', 'wow', 'hate']
```

```
def select_words(word_count):
  selected_words = ['awesome', 'great', 'fantastic', 'amazing', 'love', 'horrible', 'bad', 'terrible', 'awful', 'wow', 'hate']
  b = {}
  for word in selected_words:
    if word in word_count:
      b[word] = word_count[word]
    else:
      b[word] = 0
  return b

pd['awesome'] = pd['word_count'].apply(select_words)
pd
```

Code is used to Count the word frequency

```
def new_word_count(words):
  wordfreq = {}
  for word in words.replace(',', ' ').split():
    wordfreq[word] = wordfreq.setdefault(word, 0.0) + 1.0
  return wordfreq


```

```
max_count, min_count = 0, 9223372036854775807
max_word, min_word = selected_words[0], selected_words[0]
for word in selected_words:
    total = pd[word].sum()
    if total > max_count:
        max_word = word
        max_count = total
    if total < min_count:
        min_word = word
        min_count = total
print('most used: ', max_word, max_count)
print('least used: ', min_word, min_count)
```
