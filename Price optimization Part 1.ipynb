```
import re
import turicreate as tc
import pandas as pd
from pandas import Series,DataFrame
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams
import string
```

```
fashion_data = tc.SFrame.read_json('/content/drive/My Drive/meta_AMAZON_FASHION.json',orient='lines')
```

```
def remove_punctuation(text):
    translator = text.maketrans('', '',string.punctuation)
    text = text.translate(translator)
    return text
def average(text):
  if text == '':
    return None
  elif bool(re.search(r"\s", text)) == False:
    return float(text[:-2] +'.'+text[-2:])
  elif len(text) > 30:
    return None
  else:
    a,b = text.split("  ")
    return (int(float(a[:-2] +'.'+a[-2:])) + int(float(b[:-2] +'.'+b[-2:]))) / 2
```

```
fashion_data['price'] = fashion_data['price'].apply(remove_punctuation)
fashion_data['price'] = fashion_data['price'].apply(average)
```

```
fashion_data = fashion_data[fashion_data['price'] != None]
fashion_data = fashion_data[fashion_data['description'] != []]
```

```
fashion_data.remove_columns(['tech1','also_buy','also_view','feature','image','tech2','category','similar_item','details','main_cat','date','fit'],inplace = True)
```

```
fashion_data.rename({'asin':'ProdID'},inplace=True)
```

```
fashion = tc.SFrame.read_csv('/content/drive/My Drive/AMAZON_FASHION.csv',header = None)
```

```
fashion.rename({'X1':'ProdID','X2':'UserID','X3':'Rating','X4':'Timestamp'},inplace=True)
fashion
```

```
fashion_data = fashion_data.drop_duplicates(['ProdID'])
```

```
def Calculate_rating(asin):
  return fashion[fashion['ProdID'] == asin]['Rating'].mean()
```

```
fashion_data['rating'] = fashion_data['ProdID'].apply(Calculate_rating)
```

```
fashion_data['tf_idf'] = tc.text_analytics.tf_idf(fashion_data['title'])
```

```
def Extract(text):
  if text == '[]':
    return 'missing' 
  elif len(text) < 50:
    text = text.split()
    for v in text[1:]:
      if v != 'in':
        if v.translate(str.maketrans('', '', string.punctuation)) != '':
          if '(' in v:
              break
          value= v.translate(str.maketrans('', '', string.punctuation))
          fashion_data[value] = 1
    return text[0].translate(str.maketrans('', '', string.punctuation))
  else:
    data = fashion_data[fashion_data['rank'] == text]
    text = data['rank'].astype(list)
    text = text[0][0].split()
    for y in text[1:]:
      if y != 'in':
        if y.translate(str.maketrans('', '', string.punctuation)) != '':
          if '(' in y:
            break
          value = y.translate(str.maketrans('', '', string.punctuation))
          fashion_data[value] = 1
    return text[0].translate(str.maketrans('', '', string.punctuation))
    
```

```
fashion_data['Rank'] = fashion_data['rank'].apply(Extract)
```

```
def Filter(asin):
  df = fashion_data[fashion_data['ProdID'] == asin]
  text = df['rank']
  if text[0] == '[]':
    return 1
  elif len(text[0]) < 50:
    text = text[0].split()
    for v in text[1:]:
      if v.translate(str.maketrans('', '', string.punctuation)) == i:
        return 1
    return 0
  else:
    text = df['rank'].astype(list)
    text = text[0][0].split()
    for y in text[1:]:
      if y.translate(str.maketrans('', '', string.punctuation)) == i:
        return 1
    return 0





```

```
for i in subset_column:
  fashion_data[i] = fashion_data['ProdID'].apply(Filter)
```
